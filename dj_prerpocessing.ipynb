```python
def extract_english_words(text):
    # 영어 문자(A-Z, a-z)로 시작하거나 끝나는 부분을 찾아서 추출
    english_words = re.findall(r'[A-Za-z]+(?:[-_/\\]*[A-Za-z]+)*', text)
    return english_words

```

```python
def replace_text(text, replacements):
    # replacements 딕셔너리의 각 패턴을 텍스트에서 대체
    for old, new in replacements.items():
        text = re.sub(re.escape(old), new, text)  # re.escape를 사용하여 특수 문자도 처리
    return text

# 'dialogue'와 'summary' 열의 텍스트를 대체하는 함수
def replace_in_dataframe(df, replacements):
    # 'dialogue'와 'summary' 열에 대해 대체 작업 수행
    df['dialogue'] = df['dialogue'].apply(lambda x: replace_text(x, replacements))
    df['summary'] = df['summary'].apply(lambda x: replace_text(x, replacements))
    return df

# 'train' 데이터프레임에서 'dialogue'와 'summary' 열에 대한 대체 수행
train = replace_in_dataframe(train, replacements)

# 결과 확인
print(train.head())
```
