```python
import pandas as pd
import os
import platform

def extract_short_sentences_from_file(file_path, size_min, size_max):
    df = pd.read_csv(file_path)
    
    short_texts = []
    for index, row in df.iterrows():
        fname = row['fname']
        text = row['dialogue']
        
        dialogues = text.split('\n')
        for dialogue in dialogues:
            if size_min <= len(dialogue.strip()) <= size_max:
                short_texts.append(f"{fname}: {dialogue.strip()}")
    
    return short_texts

if __name__ == "__main__":
    os_name = platform.system()
    if os_name == 'Windows':
        PRE_PATH = ''
    elif os_name == 'Linux':
        PRE_PATH = '/kkh/'
    elif os_name == 'Darwin': # 맥
        PRE_PATH = '/kkh/'

    DATA_PATH = PRE_PATH + "data/"
    file_path = DATA_PATH + 'train.csv'
    
    # 찾으려는 길이
    size_min = 0
    size_max = 10
    short_texts = extract_short_sentences_from_file(file_path, size_min, size_max)
    for text in short_texts:
        print(text)
```
